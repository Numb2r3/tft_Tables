### take results from this tournament

```python
table = 'lobby'
sql_query = f''' 
            SELECT round,party,winner,runnerup,third,fourth,fifth,sixth,seventh,eighth FROM {table}
            WHERE tournament_id = {alternative_name.loc[0]['tournament_id']}
            ''' 
tournament_result = pd.DataFrame(sql_int.get_data_alchemy(sql_query))
tournament_result.party = tournament_result.party.astype("category")
tournament_result.party = tournament_result.party.cat.set_categories(sorter)
```

```python
def player_results(dataframe,rounds,participant,file):
    #with open("ausgabe_file.txt","a") as o:
        for i in range(participant): 
            file.write(f'|p{i+1}player={dataframe.iloc[i,0]} p{i+1}results=')
            for j in range(rounds-1):
                file.write(f'{dataframe.iloc[i,j+1]},')
            file.write(f'{dataframe.iloc[i,j+2]}\n')

def tiebreaker_result(tiebreaker,file):
    for tie in range(len(tiebreaker)):
            file.write(f'|tiebreaker{tie+1}={tiebreaker[tie]} ')    
    file.write('\n')

def promoted_results(next_round,participant,file):
    for go_up in range(next_round):
        file.write(f'|bg{go_up+1}=up')
    file.write("\n")

    for go_down in range(participant-next_round):
        file.write(f'|bg{go_down+next_round+1}=down')
    file.write("\n")

def points_result(points,file):
    temp =0
    for key in points:
        temp +=1
        file.write(f'|p{temp}={points[key]}')    
    file.write("\n")
```

```python
def liquipedia_result(dataframe,next_round,points,tiebreaker):
    rounds = len(dataframe.columns)-1
    participant = len(dataframe)
    
    with open("ausgabe_file.txt","w") as o:
        tiebreaker_result(tiebreaker,o)
        points_result(points,o)
        promoted_results(next_round,participant,o)
        player_results(dataframe,rounds,participant,o)
        
```

```python
rounds = [1,1,2,2]
groups = [32,16,8,4]
```

```python
def tabs_rounds(rounds,file):
    file.write(f'{{{{Tabs dynamic\n')
    for i in range(1,len(rounds)+1):
        file.write(f'|name{i}=Round {i} ')    
        file.write('\n')
    file.write(f'|This=1\n|hide-showall=true}}}}\n')
```

```python
def group_result(df_groups,groups,rounds,file):
    tournament_round = 0
    played = 1
    first_group = 1
    for round in rounds:
        tournament_round +=1
        file.write(f'{{{{Tabs dynamic/tab|{tournament_round}}}}} \n')
        file.write(f'==={{{{Stage|Round {tournament_round}}}}}=== \n')

        groups = df_groups.query('round == @played ').party.unique()
        
        for group in groups:
            if first_group == 1:
               file.write(f'{{{{box|start|padding=2em|padding-bottom=2em|box-width=320px}}}} \n')
            else:
                file.write(f'{{{{box|break|padding=2em|padding-bottom=2em|box-width=320px}}}} \n')

            first_group = 0 
            
            group_result = df_groups.query('party == @group')
            file.write(f'{{{{match standings|title=Group {group}|rounds={round} \n')

            points_result(dict_points,file)
            promoted_results(4,8,file)
            tiebreaker_result(tiebreaker,file)
            player_result = group_result.melt(id_vars=['round','party'],value_vars=[1,2,3,4,5,6,7,8])
            group_pivot = pd.pivot_table(player_result,values='variable',index ='value',columns='round',aggfunc= 'first').reset_index()
            for i in range(len(group_pivot)): 
                file.write(f'|p{i+1}player={group_pivot.iloc[i,0]} |p{i+1}results=')
                for j in range(round):
                    file.write(f'{group_pivot.iloc[i,j+1]},')
                file.write('\n') 
            file.write(f'}}}} \n') 
        file.write(f'{{{{box|end}}}} \n')     
        played += round
        first_group = 1
        
        
```
