```python
dict_points = {'winner': 10, 'runnerup': 8, 'third': 7, 'fourth': 6,'fifth': 4,'sixth': 3,'seventh': 2,'eighth': 1}
dict_placements = {'winner': 1, 'runnerup': 2, 'third': 3, 'fourth': 4,'fifth': 5,'sixth': 6,'seventh': 7,'eighth': 8}
tiebreaker = ['total','first','second']
tournament_result = tournament_result.rename(columns= dict_placements)
```

```python
tournament_result2 = tournament_result.melt(id_vars=['round','party'],value_vars=[1,2,3,4,5,6,7,8])
```

```python
tournament_pivot = pd.pivot_table(tournament_result2,values='variable',index ='value',columns='round',aggfunc= 'first').reset_index()
```

```python
tournament_pivot
```

```python
with open("ausgabe_file.txt","w") as o:
    for i in range(len(tournament_pivot)): 
        o.write(f'|p{i+1}player={tournament_pivot.iloc[i,0]} p{i+1}results=')
        for j in range(len(tournament_pivot.columns)-2):
             o.write(f'{tournament_pivot.iloc[i,j+1]},')
        o.write(f'{tournament_pivot.iloc[i,j+2]}\n')
```

```python
def player_results(dataframe,rounds,participant,file):
    #with open("ausgabe_file.txt","a") as o:
        for i in range(participant): 
            print(str(i))
            file.write(f'|p{i+1}player={dataframe.iloc[i,0]} p{i+1}results=')
            for j in range(rounds-1):
                file.write(f'{dataframe.iloc[i,j+1]},')
            file.write(f'{dataframe.iloc[i,j+2]}\n')

def tiebreaker_result(tiebreaker,file):
    for tie in range(len(tiebreaker)):
            file.write(f'|tiebreaker{tie+1}={tiebreaker[tie]} ')    
    file.write('\n')

def promoted_results(next_round,participant,file):
    for go_up in range(next_round):
        file.write(f'|bg{go_up+1}=up')
    file.write("\n")

    for go_down in range(participant-next_round):
        file.write(f'|bg{go_down+next_round+1}=down')
    file.write("\n")

def points_result(points,file):
    temp =0
    for key in points:
        temp +=1
        file.write(f'|p{temp}={points[key]}')    
    file.write("\n")
```

```python
def liquipedia_result(dataframe,next_round,points,tiebreaker):
    rounds = len(dataframe.columns)-1
    participant = len(dataframe)
    
    with open("ausgabe_file.txt","w") as o:
        tiebreaker_result(tiebreaker,o)
        points_result(points,o)
        promoted_results(next_round,participant,o)
        player_results(dataframe,rounds,participant,o)
        
```
